```python
# Salin dataframe agar data asli tetap utuh
df_clean = df.copy()

# Pisahkan fitur dan target
target = 'loan_status'
X = df_clean.drop(columns=[target])
y = df_clean[target]

# Identifikasi tipe fitur
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

numerical_features, categorical_features
```

```python
# Cek missing values
df_clean.isnull().sum()
```

```python
# Imputer
numeric_imputer = SimpleImputer(strategy='median')
categorical_imputer = SimpleImputer(strategy='most_frequent')

# Terapkan imputasi
X[numerical_features] = numeric_imputer.fit_transform(X[numerical_features])
X[categorical_features] = categorical_imputer.fit_transform(X[categorical_features])

# Verifikasi
X.isnull().sum()
```

```python
# Pipeline numerik
numeric_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Pipeline kategorikal
categorical_pipeline = Pipeline(steps=[
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Gabungkan preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)

# Fit dan transform data
X_processed = preprocessor.fit_transform(X)

X_processed
```

```python
# Simpan preprocessor
joblib.dump(preprocessor, 'preprocessor.joblib')

# Simpan dataset hasil preprocessing
joblib.dump((X_processed, y), 'dataset_preprocessed.joblib')
```
